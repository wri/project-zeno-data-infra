```python
base_path = "s3://gfw-data-lake/umd_glad_dist_alerts/parquet2/"
tables = ["gadm_dist_alerts", "gadm_dist_alerts_by_driver", "gadm_dist_alerts_by_natural_lands", "gadm_dist_alerts_by_grasslands", "gadm_dist_alerts_by_land_cover"]
suffixes = ["", "-by-driver", "by-natural-land-class", "-by-grassland-class", "-by-land-cover-class"]

for table, suffix in zip(tables, suffixes):
    df = pd.read_parquet(
        f"{base_path}{table}.parquet",
        storage_options={
            "requester_pays": True,  # alias for RequestPayer in fsspec
        }
    )
    renamed_df = df.rename(columns={
        "area__ha": "area_ha",
        "alert_date": "dist_alert_date",
        "alert_confidence": "dist_alert_confidence"
    })

    renamed_df["aoi_type"] = "admin"
    renamed_df["area_ha"] = renamed_df["area_ha"] / 10000
    
    renamed_df.to_parquet(f"s3://lcl-analytics/zonal-statistics/admin-dist-alerts{suffix}.parquet", index=False)
```
